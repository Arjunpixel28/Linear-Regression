Handling the Duplicates

```python
data.duplicated().sum() 
data = data.drop_duplicates()
```

Selecting features and target variable

```python
X = data[['Year', 'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure', 'Hepatitis B', 'Measles ', ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure', 'Diphtheria ', ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years', ' thinness 5-9 years', 'Income composition of resources', 'Schooling']]
y = data['Life expectancy ']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print("R-squared:", r2)
```

```python


# Scatter plot
plt.scatter(y_test, y_pred)

# Add a line
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='-')

# Labels and title
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.title('Predicted vs Actual Life Expectancy')

# Show the plot
plt.show()
```
